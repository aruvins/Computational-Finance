```python
LLY_ = LLY.copy()
LLY_ = LLY_.drop(columns=['open', 'high', 'low', 'close', 'dividends', 'splits'])
LLY_ = LLY_.rename(columns={'adjclose': 'LLY_adj', 'volume': 'LLY_volume'})
LLY_.info()
LLY_.head()
```

```python
SPY_ = SPY.copy()
SPY_ = SPY_.drop(columns=['open', 'high', 'low', 'close', 'dividends', 'splits'])
SPY_ = SPY_.rename(columns={'adjclose': 'SPY_adj', 'volume': 'SPY_volume'})
SPY_.info()
SPY_.head()
```

```python
SPY_LLY = pd.merge(LLY_, SPY_, on='date', how='inner')
SPY_LLY.info()
SPY_LLY.head()
volRatio = SPY_LLY[SPY_LLY['date'] == '2015-12-30']
volRatio['volume_ratio'] = volRatio['LLY_volume'] / volRatio['SPY_volume']
volRatio
```

```python
SPY_LLY['LLY_adj'].pct_change()*100
```

```python
SPY_LLY['LLY_adj'].pct_change()*100
```

```python
returns = pd.DataFrame({
    'LLY_ret': SPY_LLY['LLY_adj'].pct_change(),
    'LLY_next_ret': SPY_LLY['LLY_adj'].pct_change().shift(-1),
    'SPY_ret': SPY_LLY['SPY_adj'].pct_change(),
    'SPY_next_ret': SPY_LLY['SPY_adj'].pct_change().shift(-1),
    'LLY_volume': SPY_LLY['LLY_volume'],
    'SPY_volume': SPY_LLY['SPY_volume']
})

# Number of missing values in each column
returns.isnull().sum()
```

```python
returns_ = returns.dropna()
returns_descriptive = returns_.describe()

highest = returns_descriptive.loc['max', 'LLY_ret']
lowest = returns_descriptive.loc['min', 'LLY_ret']

print("Highest LLY_ret:", highest)
print("Lowest LLY_ret:", lowest)
```

```python
returns_ = returns.dropna()
highest_ret_day = returns_[returns_['LLY_ret'] == returns_['LLY_ret'].max()]
highest_vol_day = returns_[returns_['LLY_volume'] == returns_['LLY_volume'].max()]

# Median LLY_volume
median_vol = returns_['LLY_volume'].median()

highest_ret_day
```
